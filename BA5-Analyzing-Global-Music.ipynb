### **4.3.5 Converting the spaces in Artists with an `_` (data consistency)**

```python
# Replace whitespace in "artist" column with underscore
from pyspark.sql.functions import regexp_replace

spotify_df = spotify_df.withColumn("artist", regexp_replace("artist", " ", "_"))
```

The code above replaces any white spaces in the "artist" column of the Spotify dataframe with underscores. This is a common preprocessing step when dealing with data that may be used for further analysis or modeling, as it can make it easier to work with the data. For example, **Billie Eilish** if left with a space could lead to results where `Billie` and `Eilish` are showing up separately, which we do not want.

### **4.3.6 Converting the spaces in Regions with an `_` (data consistency)**

```python
import pyspark.sql.functions as F
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

# Define a UDF to clean the region text
def clean_region(text):
    # Remove links from text
    text = re.sub(r'http\S+|www.\S+', '', str(text))
    # Remove leading/trailing spaces
    text = text.strip()
    # Replace spaces with underscores
    text = text.replace(' ', '_')
    return text

# Define a UDF for cleaning the region column
clean_region_udf = udf(clean_region, StringType())

# Apply the clean_region_udf to the region column
spotify_df = spotify_df.withColumn('region', clean_region_udf('region'))

# Drop rows where region is null, NaN, or empty
spotify_df = spotify_df.filter(spotify_df.region.isNotNull() & (spotify_df.region != ''))
```

This code defines a function called clean_region() that cleans the text in the region column by **removing links, leading/trailing spaces, and replacing spaces with underscores**. 

Then it defines another function called `clean_region_udf` that applies the `clean_region()` function to the `region` column of the `spotify_df` dataframe.

Finally, it removes any rows in the `spotify_df` dataframe where the region column is `null`,`NaN`, or `empty`.

## **5.2 World map of the most streams per region**

```python
%%capture
!pip install geopandas
!pip install geoplot
!pip install plotly
!pip install pycountry
```

```python
import geopandas as gpd
import pycountry

# Aggregate the data by region and sum the total streams
region_streams = spotify_df.groupby('region').agg({'Streams': 'sum'}).toPandas()

# Remove 'Global' region
region_streams = region_streams[region_streams['region'] != 'Global']

# Convert scientific notation to regular numbers
region_streams['sum(Streams)'] = region_streams['sum(Streams)'].apply(lambda x: '{:.0f}'.format(x))

# Remove rows with 'nan' values in sum(Streams) column
region_streams = region_streams[~region_streams['sum(Streams)'].astype(str).str.contains('nan', flags=re.IGNORECASE)]

# Manually match regions to iso codes
manual_match = {'South_Africa': 'ZAF',
                'New_Zealand': 'NZL',
                'Czech_Republic': 'CZE',
                'United_States': 'USA',
                'United_Arab_Emirates': 'ARE',
                'Saudi_Arabia': 'SAU',
                'Hong_Kong': 'HKG',
                'United_Kingdom': 'GBR',
                'El_Salvador': 'SLV',
                'Dominican_Republic': 'DOM',
                'Costa_Rica': 'CRI',
                'South_Korea': 'KOR'}

# Add the iso_code column to the region_streams DataFrame
region_streams['iso_code'] = np.nan
region_streams['sum(Streams)'] = region_streams['sum(Streams)'].astype(int)

# Fill in the iso_code column where it's empty
for region in region_streams['region']:
    if pd.isna(region_streams.loc[region_streams['region'] == region, 'iso_code'].iloc[0]):
        if region in manual_match:
            region_streams.loc[region_streams['region'] == region, 'iso_code'] = manual_match[region]
        else:
            try:
                country = pycountry.countries.search_fuzzy(region)[0]
                region_streams.loc[region_streams['region'] == region, 'iso_code'] = country.alpha_3
            except LookupError:
                print(f"No match found for {region}")

region_streams.to_csv('region_streams.csv', index=False)
```

```python
# Load the world dataset from geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Rename the iso_a3 column to iso_code
world = world.rename(columns={'iso_a3': 'iso_code'})

# Merge the region_streams data frame with the GeoDataFrame using iso_code column
merged_data = world.merge(region_streams, on='iso_code', how='left')

# Remove NaN rows from merged_data
merged_data = merged_data.dropna(subset=['sum(Streams)'])

# Convert sum(Streams) column to numeric
merged_data['sum(Streams)'] = pd.to_numeric(merged_data['sum(Streams)'])

# Format sum(Streams) values
merged_data['sum(Streams)'] = merged_data['sum(Streams)'].apply(lambda x: '{:.1f}'.format(x/1e9))

# Convert sum(Streams) column to float
merged_data['sum(Streams)'] = merged_data['sum(Streams)'].astype(float)

# Sort merged_data by descending order of sum(Streams)
merged_data = merged_data.sort_values(by='sum(Streams)', ascending=False)

# Plot the map
fig, ax = plt.subplots(figsize=(12, 8))

# Set the background color
ax.set_facecolor('#FFFFFF')

# Set the facecolor of the merged_data GeoDataFrame to white
merged_data.plot(column='sum(Streams)', cmap='Greens', linewidth=0.8, ax=ax, edgecolor='black', legend=True,
                 legend_kwds={'label': 'Streams in (billions)', 'orientation': 'horizontal', 'format': '%.0f'},
                 vmin=0, vmax=35)

ax.tick_params(colors='black')

# Set the title and font size
ax.set_title('Total Streams by Country', fontsize=15)

plt.show()
```

## **5.3 Songs Longevity that ranked 1st**

```python
# Convert Spark DataFrame to Pandas DataFrame
spotify_df.createOrReplaceTempView('spotify')
songs_longevity = spark.sql('''
SELECT title, count(date) AS Days
FROM spotify
WHERE chart='top200'
AND rank=1
AND region='United_States'
GROUP BY title
ORDER BY Days DESC
LIMIT 10;
''').toPandas()

# Create a horizontal bar chart
fig, ax = plt.subplots(figsize=(12, 8))
ax.barh(songs_longevity['title'].sort_values(ascending=True), songs_longevity['Days'].sort_values(ascending=True), color='#1DB954')
ax.set_xlabel('Days')
ax.set_ylabel('Title')

# Set the title of the chart
plt.title('Top 10 Songs with Longest Longevity on US Top 200 Chart')

# Display the chart
plt.show()
```

The summary highlights the **top ten most popular songs in the United States over a certain period, based on their longevity at the top of the charts**.

The summary highlights the importance of longevity in the music industry. The fact that the top songs have remained popular for an extended period indicates their **lasting appeal and potential to generate revenue for music industry investors**. This could suggest a shift in the music industry, with **more emphasis placed on creating long-lasting hits as opposed to one-off popular songs**.

### 5.8.4 K-Means clustering

The idea behind K-Means is in grouping the data into clusters such that the variation inside the clusters (also known as total within-cluster sum of square or WSS) is minimum, and the variation within the clusters is maximum.

This helps in understanding which songs tend to be popular in which groups.

The process in the code involves looping through different values of K (the number of clusters) and fitting a KMeans model to the data using that value of K. For each value of K, the within-cluster sum of squares (WSS) is calculated, which is a measure of how well the data points within each cluster are grouped together. The WSS is plotted against the number of clusters to help identify the optimal value of K.



```python
from sklearn.cluster import KMeans

wss = []
for i in range(1, 16):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(spotify_train)
    wss.append(kmeans.inertia_)
    
plt.plot(range(1, 16), wss, '-o', color='#1DB954')
plt.xlabel('Number of Clusters K')
plt.ylabel('Total WSS')
plt.title('Optimal Number of Clusters')
plt.show()
```

The optimal value of K is often chosen as the "elbow point" of the WSS curve, which is the point at which the rate of decrease in WSS starts to level off. This point represents a balance between maximizing the similarity of data points within each cluster and minimizing the number of clusters used.

With the elbow method, the ideal number of clusters are identified as 2. Therefore kmeans is implemented with 2 centers.

**Fitting the K Means Model**

The k means model is fit with 2 centers.

As seen from the output this model results in 2 clusters of sizes 190, 506.

```python
from sklearn.cluster import KMeans

# Fit k-means with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42).fit(spotify_train)

# Get cluster sizes
cluster_sizes = pd.Series(kmeans.labels_).value_counts().sort_index()

print(cluster_sizes)
```

**Silhouette Score**

The Silhouette Score ranges from -1 to 1, where a score of 1 indicates that the clusters are well separated, and a score of -1 indicates that the clusters are overlapping. A score close to 0 indicates that the clusters are close to each other.

We tend to get the highest silhouette_score for k=2, hence we pick the ideal number of clusters to be 2. 

```python
from sklearn.metrics import silhouette_score

# Calculate Silhouette Score
silhouette_score = silhouette_score(spotify_train, train_data['cluster'], metric='euclidean')
#print("Silhouette Score:", silhouette_score)
print("Silhouette Score: {:.5f}".format(silhouette_score))
```
